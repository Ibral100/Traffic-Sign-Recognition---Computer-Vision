```python
import pandas as pd
import cv2
import numpy as np
import os

# recuperer les rois de toutes les images 4



path_train_csv = "archive/Train.csv"

df_train = pd.read_csv(path_train_csv)

print(df_train.head())
```

```python
# definir une taille fixe des rois 


for index,row in df_train.iterrows():

    x1, y1, x2, y2 = int(row["Roi.X1"]), int(row["Roi.Y1"]), int(row["Roi.X2"]), int(row["Roi.Y2"])

    df_train.at[index,"Roi_height"] = y2-y1+1
    df_train.at[index,"Roi_width"] = x2-x1+1

    #print(df_train.at[index,"Roi_height"])
    #print(df_train.at[index,"Roi_width"])



```

```python
print(df_train["Roi_height"].max())
print(df_train["Roi_width"].max())

print(df_train["Roi_height"].min())
print(df_train["Roi_width"].min())

print(df_train.describe())
```

```python
## parcourir toutes les images et recuperer uniquement les pixels dans le roi
# definir la taille fixe a 64,64 


for index, row in df_train.iterrows():
    
    img_path = "archive/" + row["Path"]
    img = cv2.imread(img_path)
    

    
    x1, y1, x2, y2 = int(row["Roi.X1"]), int(row["Roi.Y1"]), int(row["Roi.X2"]), int(row["Roi.Y2"])

    img_roi = img[y1:y2, x1:x2]

    h=img_roi.shape[0]
    w=img_roi.shape[1]

    cote = max(h,w)

    carre_noir = np.zeros(shape=(cote,cote,3),dtype=np.uint8)

    x_offset = (cote-w) // 2
    y_offset = (cote-h) // 2

    carre_noir[y_offset:y_offset+h,x_offset:x_offset+w] = img_roi

    img_finale =  cv2.resize(carre_noir,(64,64),interpolation=cv2.INTER_AREA)

    path_save = "archive/Train_Roi/" + row["Path"].replace("Train/", "")
    
    os.makedirs(os.path.dirname(path_save), exist_ok=True)

    cv2.imwrite(path_save, img_finale)




```
